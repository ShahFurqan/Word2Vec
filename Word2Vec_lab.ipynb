```python
import nltk
nltk.download("punkt")
nltk.download("stopwords")
from gensim.models import Word2Vec
from nltk.corpus import stopwords
#re is real grass version 
import re
```

```python
paragraph="A paragraph is a self-contained unit of discourse in writing dealing with a particular point or idea. Though not required by the orthographic conventions of any language with a 222 .The writing system, paragraphs are a conventional means of organizing extended segments"
```

```python
# pre-processing the data 
text = re.sub(r'\[[0-9]*\]',' ',paragraph) # removing any no in paragraph and passing the paragraph 
text=re.sub(r'\s+',' ',text)    # removing any special charactre in paragraph 
text=text.lower()
text=re.sub(r'\d',' ',text) # a substring and returns a string with replaced values
text = re.sub(r'\s+',' ',text)
```

```python
# preparing the data set
sentences= nltk.sent_tokenize(text)
sentence =[nltk.word_tokenize(sentence) for sentence in sentences]

for i in range (len(sentences)):
    sentence [i] = [word for word in sentences[i] if word not in stopwords.words('english') ] # checking not any stop word

```

```python
# training the word2vec model 
model = Word2Vec(sentences,min_count=1) # minimun count 1
```

```python
vectors = model.wv.vectors # for 2D table I saw in lecture
print(vectors)
```

```python
keys = model.wv.vocab.keys() 
# keys
for i,j in zip (keys,vectors): # here keys are just individual words
    print(i,':',j)
```
